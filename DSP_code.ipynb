```python
span_pattern = (
    r"span id:([A-Za-z0-9_]+) "
    r"annotator:([A-Za-z0-9_]+) "
    r"layer:([A-Za-z0-9_]+) "
    r"feature:([A-Za-z0-9_]+) "
    r"tag:([A-Za-z0-9_]+) "
    r"start:([0-9]+) "
    r"stop:([0-9]+) "
    r"text:([^,]+)"
)

def extract_annotations(row):
    curation_list = row['Curation']
    policy = row['Policy']
    text = row['Text']
    spans = []

    for span_text in curation_list:

        span_text = str(span_text)

        match = re.search(span_pattern, span_text)

        if match:
            span_id, annotator, layer, feature, tag, start, stop, annotation_text = match.groups()

            spans.append({
                'Policy': policy,
                'Text': text,
                'Span_ID': span_id,
                'Annotator': annotator,
                'Layer': layer,
                'Feature': feature,
                'Tag': tag,
                'Start': int(start),
                'Stop': int(stop),
                'Annotation_Text': annotation_text.strip()
            })

    return spans

all_annotations = df.apply(extract_annotations, axis=1)

flat_annotations = [annotation for annotations in all_annotations for annotation in annotations]

annotations_df = pd.DataFrame(flat_annotations)

annotations_df
```

# Sampling

```python
import os
import numpy as np
import pandas as pd
import logging

from imblearn.over_sampling import ADASYN, SMOTE, RandomOverSampler
from imblearn.combine import SMOTEENN
from imblearn.under_sampling import RandomUnderSampler

from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

import torch

logging.basicConfig(filename="resampling_log.txt", level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

logging.info(">>> Start des Resampling-Prozesses <<<")

data = pd.read_csv("annotations.csv")
logging.info("Datei erfolgreich geladen")

X_text = data["Text"]
y = data["Layer"]

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
logging.info("Label-Encoding abgeschlossen")

vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(X_text)
logging.info("TF-IDF-Vektorisierung abgeschlossen.")

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)
logging.info(f"Percentage Split abgeschlossen. Trainingsgröße: {len(y_train)}, Testgröße: {len(y_test)}")

output_dir = "resampled_datasets"
os.makedirs(output_dir, exist_ok=True)
logging.info("Hauptverzeichnis '{output_dir}' für Resampling-Daten erstellt")

SAMPLERS = {
    "Adasyn": ADASYN(random_state = 42),
    "SMOTE": SMOTE(random_state = 42),
    "SMOTEENN": SMOTEENN(random_state = 42),
    "Oversampling": RandomOverSampler(random_state = 42),
    "Undersampling": RandomUnderSampler(random_state = 42),
}

resampled_distribution = {}

for strategy, sampler in SAMPLERS.items():
    logging.info(f"Beginne Resampling mit Methode: {strategy}")
    X_train_resampled, y_train_resampled = sampler.fit_resample(X_train, y_train)
    logging.info(f"Resampling mit {strategy} abgeschlossen. Neue Größe: {len(y_train_resampled)}")

    strategy_folder = os.path.join(output_dir, strategy)
    os.makedirs(strategy_folder, exist_ok=True)
    logging.info(f"Unterverzeichnis '{strategy_folder}' erstellt")

    X_train_resampled_text = vectorizer.inverse_transform(X_train_resampled)
    resampled_df = pd.DataFrame({"Text": [" ".join(words) for words in X_train_resampled_text], "Layer": y_train_resampled})
    file_path = os.path.join(strategy_folder, f"{strategy}_resampled.csv")
    resampled_df.to_csv(file_path, index=False)
    logging.info(f"Resampled-Dataset gespeichert: {file_path}")

    resampled_distribution[strategy] = np.bincount(y_train_resampled)

resampled_distribution_df = pd.DataFrame(resampled_distribution)
logging.info("DataFrame aus den gesammelten Resampling-Informationen wurde erstellt")

resampled_distribution_df.index = label_encoder.classes_
logging.info("Klassen werden als Index gesetzt")

resampled_distribution_path = os.path.join(output_dir, "resampled_class_distributions.csv")
resampled_distribution_df.to_csv(resampled_distribution_path)
logging.info("Speicherung des DataFrames als .csv-Datei")

train_counts = np.bincount(y_train)
test_counts = np.bincount(y_test)

original_distribution_df = pd.DataFrame({"Train:": train_counts, "Test:": test_counts}, index=label_encoder.classes_)
original_distribution_path = os.path.join(output_dir, "original_train_test_distribution.csv")
original_distribution_df.to_csv(original_distribution_path)

logging.info(f"Original Train-Test Verteilung gespeichert: {original_distribution_path}")
```

# Training

```python
import logging
import numpy as np
import pandas as pd
import torch
from pathlib import Path
from transformers import AutoModelForSequenceClassification, AutoTokenizer, Trainer, TrainingArguments
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import datasets
import os

os.environ["TOKENIZERS_PARALLELISM"] = "false"
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"

logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
logging.info(f"Using device: {device}")

LEARNING_RATES = [2e-5, 5e-5]
BATCH_SIZES = [8]
MODEL_NAME = "ClimateBERT/distilroberta-base-climate-f"
N_OUTER_SPLITS = 2
N_INNER_SPLITS = 2
N_EPOCHS = 2

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy_score(labels, predictions),
        "f1": f1_score(labels, predictions, average="macro"),
        "precision": precision_score(labels, predictions, average="macro"),
        "recall": recall_score(labels, predictions, average="macro"),
    }

def train_and_evaluate(train_texts, train_labels, val_texts, val_labels, model_name, learning_rate, batch_size, n_epochs):
    logging.info(f"Training with LR: {learning_rate}, Batch Size: {batch_size}")

    tokenizer = AutoTokenizer.from_pretrained(model_name)

    train_encodings = tokenizer(list(train_texts), truncation=True, padding=True, max_length=512)
    val_encodings = tokenizer(list(val_texts), truncation=True, padding=True, max_length=512)

    train_dataset = datasets.Dataset.from_dict({
        "input_ids": train_encodings["input_ids"],
        "attention_mask": train_encodings["attention_mask"],
        "labels": train_labels.to_numpy().tolist()
    })

    val_dataset = datasets.Dataset.from_dict({
        "input_ids": val_encodings["input_ids"],
        "attention_mask": val_encodings["attention_mask"],
        "labels": val_labels.to_numpy().tolist()
    })

    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=len(set(train_labels)))

    training_args = TrainingArguments(
        output_dir="./results",
        evaluation_strategy="epoch",
        per_device_train_batch_size=batch_size,
        per_device_eval_batch_size=batch_size,
        learning_rate=learning_rate,
        num_train_epochs=n_epochs,
        logging_dir="./logs",
        logging_steps=10,
        save_strategy="epoch",
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        compute_metrics=compute_metrics,
    )

    trainer.train()
    eval_results = trainer.evaluate()

    return eval_results

def process_strategy(strategy, output_folder, model_name):
    logging.info(f"Processing strategy: {strategy}")
    data_file = Path(output_folder) / strategy / f"{strategy}_resampled.csv"

    if not data_file.exists():
        logging.warning(f"Data file not found for strategy {strategy}")
        return None

    data = pd.read_csv(data_file)
    texts, labels = data['Text'], data['Layer']

    outer_cv = StratifiedKFold(n_splits=N_OUTER_SPLITS, shuffle=True, random_state=42)

    strategy_results = {
        "best_params": None,
        "best_results": None,
        "tested_combinations": []
    }

    best_f1 = 0

    for train_idx, test_idx in outer_cv.split(texts, labels):
        train_texts, test_texts = texts.iloc[train_idx], texts.iloc[test_idx]
        train_labels, test_labels = labels.iloc[train_idx], labels.iloc[test_idx]

        param_grid = {
            "learning_rate": LEARNING_RATES,
            "batch_size": BATCH_SIZES
        }

        for lr in LEARNING_RATES:
            for bs in BATCH_SIZES:
                inner_scores = []
                metrics_list = []

                inner_cv = StratifiedKFold(n_splits=N_INNER_SPLITS, shuffle=True, random_state=42)

                for inner_train_idx, val_idx in inner_cv.split(train_texts, train_labels):
                    inner_train_texts, val_texts = train_texts.iloc[inner_train_idx], train_texts.iloc[val_idx]
                    inner_train_labels, val_labels = train_labels.iloc[inner_train_idx], train_labels.iloc[val_idx]

                    eval_results = train_and_evaluate(inner_train_texts, inner_train_labels, val_texts, val_labels, model_name, lr, bs, N_EPOCHS)
                    inner_scores.append(eval_results.get("eval_f1", 0))
                    metrics_list.append(eval_results)

                if not metrics_list:
                    logging.warning("metrics_list is empty, skipping parameter evaluation.")
                    continue

                avg_metrics = {metric: np.mean([m[metric] for m in metrics_list]) for metric in metrics_list[0].keys()}

                strategy_results["tested_combinations"].append({
                    "learning_rate": lr,
                    "batch_size": bs,
                    "metrics": avg_metrics
                })

                if avg_metrics.get("eval_f1", 0) > best_f1:
                    best_f1 = avg_metrics["eval_f1"]
                    strategy_results["best_params"] = {"learning_rate": lr, "batch_size": bs}
                    strategy_results["best_results"] = avg_metrics

    logging.info(f"Best parameters for {strategy}: {strategy_results['best_params']}")

    # Ergebnisse abspeichern
    results_path = Path("results") / f"{strategy}_results.json"
    results_path.parent.mkdir(parents=True, exist_ok=True)

    pd.DataFrame(strategy_results["tested_combinations"]).to_json(results_path, orient="records", indent=4)

    return strategy_results

output_folder = Path("resampled_datasets")

strategy = "Adasyn"
result_adasyn = process_strategy(strategy, output_folder, MODEL_NAME)

strategy = "SMOTE"
result_smote = process_strategy(strategy, output_folder, MODEL_NAME)

strategy = "SMOTEENN"
result_smoteenn = process_strategy(strategy, output_folder, MODEL_NAME)

strategy = "Oversampling"
result_oversampling = process_strategy(strategy, output_folder, MODEL_NAME)

strategy = "Undersampling"
result_undersampling = process_strategy(strategy, output_folder, MODEL_NAME)
```

```python
def train_and_evaluate(train_texts, train_labels, val_texts, val_labels, model_name, learning_rate, batch_size, n_epochs):
    logging.info(f"Training with LR: {learning_rate}, Batch Size: {batch_size}")

    tokenizer = AutoTokenizer.from_pretrained(model_name)

    train_encodings = tokenizer(list(train_texts), truncation=True, padding=True, max_length=512)
    val_encodings = tokenizer(list(val_texts), truncation=True, padding=True, max_length=512)

    train_dataset = datasets.Dataset.from_dict({
        "input_ids": train_encodings["input_ids"],
        "attention_mask": train_encodings["attention_mask"],
        "labels": torch.tensor(train_labels.to_numpy()).tolist()
    })

    val_dataset = datasets.Dataset.from_dict({
        "input_ids": val_encodings["input_ids"],
        "attention_mask": val_encodings["attention_mask"],
        "labels": torch.tensor(val_labels.to_numpy()).tolist()
    })

    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=len(set(train_labels)))

    training_args = TrainingArguments(
        output_dir="./results",
        evaluation_strategy="epoch",
        per_device_train_batch_size=batch_size,
        per_device_eval_batch_size=batch_size,
        learning_rate=learning_rate,
        num_train_epochs=n_epochs,
        logging_dir="./logs",
        logging_steps=10,
        save_strategy="epoch",
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        compute_metrics=compute_metrics,
    )

    trainer.train()
    eval_results = trainer.evaluate()

    return eval_results

from sklearn.model_selection import train_test_split

def process_percentage_split(data_path, model_name):
    logging.info(f"Processing percentage split with dataset {data_path}")

    data = pd.read_csv(data_path)
    texts, labels = data["Text"], data["Layer_encoded"]


    train_texts, test_texts, train_labels, test_labels = train_test_split(
    texts, labels, test_size=0.2, stratify=labels, random_state=42)

    val_texts, test_texts, val_labels, test_labels = train_test_split(
    test_texts, test_labels, test_size=0.5, stratify=test_labels, random_state=42)


    best_f1 = 0
    best_params = None
    best_results = None
    tested_combinations = []

    for lr in LEARNING_RATES:
        for bs in BATCH_SIZES:
            eval_results = train_and_evaluate(train_texts, train_labels, val_texts, val_labels, model_name, lr, bs, N_EPOCHS)
            tested_combinations.append({
                "learning_rate": lr,
                "batch_size": bs,
                "metrics": eval_results
            })

            if eval_results.get("eval_f1", 0) > best_f1:
                best_f1 = eval_results["eval_f1"]
                best_params = {"learning_rate": lr, "batch_size": bs}
                best_results = eval_results

    logging.info(f"Best parameters: {best_params}")

    results_path = Path("results") / "percentage_split_results.json"
    results_path.parent.mkdir(parents=True, exist_ok=True)

    pd.DataFrame(tested_combinations).to_json(results_path, orient="records", indent=4)

    return {
        "best_params": best_params,
        "best_results": best_results,
        "tested_combinations": tested_combinations
    }

data_path = "annotations.csv"
result_percentage_split = process_percentage_split(data_path, MODEL_NAME)
```

```python
def retrain_and_predict(data_path, model_name, learning_rate, batch_size, output_csv_path):
    logging.info("Retraining with the best strategy and generating predictions...")

    data = pd.read_csv(data_path)
    texts, labels = data["Text"], data["Layer_encoded"]

    tokenizer = AutoTokenizer.from_pretrained(model_name)
    encodings = tokenizer(list(texts), truncation=True, padding=True, max_length=512)

    dataset = datasets.Dataset.from_dict({
        "input_ids": encodings["input_ids"],
        "attention_mask": encodings["attention_mask"],
        "labels": labels.tolist()
    })

    model = AutoModelForSequenceClassification.from_pretrained(
        model_name, num_labels=len(set(labels))
    )

    training_args = TrainingArguments(
        output_dir="./results",
        evaluation_strategy="no",
        per_device_train_batch_size=batch_size,
        learning_rate=learning_rate,
        num_train_epochs=2,
        save_strategy="no",
        logging_dir="./logs",
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=dataset,
    )

    trainer.train()

    predictions = trainer.predict(dataset)
    predicted_labels = np.argmax(predictions.predictions, axis=-1)

    data["Predicted_Layer"] = predicted_labels
    data.to_csv(output_csv_path, index=False)

    logging.info(f"Predictions saved to {output_csv_path}")

data_path = "annotations.csv"
output_csv_path = "annotations_with_predictions.csv"
retrain_and_predict(data_path, MODEL_NAME, 5e-5, 8, output_csv_path)
```

```python
def train_and_predict_feature_with_layer(data_path, model_name, learning_rate, batch_size, output_csv_path):
    logging.info("Training to predict Feature using Text and Predicted_Layer...")

    data = pd.read_csv(data_path)
    texts = data["Text"]
    predicted_layer = data["Predicted_Layer"]
    target_feature = data["Feature"]

    tokenizer = AutoTokenizer.from_pretrained(model_name)
    encodings = tokenizer(
        list(texts), truncation=True, padding=True, max_length=512
    )

    input_features = {
        "input_ids": encodings["input_ids"],
        "attention_mask": encodings["attention_mask"],
        "predicted_layer": predicted_layer.tolist()
    }

    dataset = datasets.Dataset.from_dict({
        **input_features,
        "labels": target_feature.astype("category").cat.codes.tolist()
    })

    num_labels = len(data["Feature"].unique())
    model = AutoModelForSequenceClassification.from_pretrained(
        model_name, num_labels=num_labels
    )

    training_args = TrainingArguments(
        output_dir="./results_with_layer",
        evaluation_strategy="no",
        per_device_train_batch_size=batch_size,
        learning_rate=learning_rate,
        num_train_epochs=2,
        save_strategy="no",
        logging_dir="./logs_with_layer",
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=dataset,
    )

    trainer.train()

    predictions = trainer.predict(dataset)
    predicted_features = np.argmax(predictions.predictions, axis=-1)

    data["Predicted_Feature_With_Layer"] = predicted_features
    data.to_csv(output_csv_path, index=False)

    logging.info(f"Predictions with Predicted_Layer saved to {output_csv_path}")


def train_and_predict_feature_without_layer(data_path, model_name, learning_rate, batch_size, output_csv_path):
    logging.info("Training to predict Feature using only Text...")

    data = pd.read_csv(data_path)
    texts = data["Text"]
    target_feature = data["Feature"]

    tokenizer = AutoTokenizer.from_pretrained(model_name)
    encodings = tokenizer(
        list(texts), truncation=True, padding=True, max_length=512
    )

    dataset = datasets.Dataset.from_dict({
        "input_ids": encodings["input_ids"],
        "attention_mask": encodings["attention_mask"],
        "labels": target_feature.astype("category").cat.codes.tolist()
    })

    num_labels = len(data["Feature"].unique())
    model = AutoModelForSequenceClassification.from_pretrained(
        model_name, num_labels=num_labels
    )

    training_args = TrainingArguments(
        output_dir="./results_without_layer",
        evaluation_strategy="no",
        per_device_train_batch_size=batch_size,
        learning_rate=learning_rate,
        num_train_epochs=2,
        save_strategy="no",
        logging_dir="./logs_without_layer",
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=dataset,
    )

    trainer.train()

    predictions = trainer.predict(dataset)
    predicted_features = np.argmax(predictions.predictions, axis=-1)

    data["Predicted_Feature_Without_Layer"] = predicted_features
    data.to_csv(output_csv_path, index=False)

    logging.info(f"Predictions without Layer saved to {output_csv_path}")

data_path = "annotations_with_predictions.csv"
output_csv_with_layer = "annotations_with_predictions_with_feature_with_layer.csv"
output_csv_without_layer = "annotations_with_predictions_with_feature_without_layer.csv"

train_and_predict_feature_with_layer(
    data_path, MODEL_NAME, 5e-5, 8, output_csv_with_layer
)

train_and_predict_feature_without_layer(
    data_path, MODEL_NAME, 5e-5, 8, output_csv_without_layer
)
```

```python
import os
import numpy as np
import pandas as pd
import logging

from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

import torch

logging.basicConfig(filename="resampling_log_feature.txt", level=logging.INFO,
                    format="%(asctime)s - %(levelname)s - %(message)s")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

logging.info(">>> Start des Resampling-Prozesses für Feature Prediction <<<")

data_path = "annotations_with_predictions.csv"
data = pd.read_csv(data_path)
logging.info(f"Datei erfolgreich geladen: {data_path}")


X_text = data["Text"]
X_layer = data["Predicted_Layer"]
y_feature = data["Feature"]

feature_encoder = LabelEncoder()
y_feature_encoded = feature_encoder.fit_transform(y_feature)

logging.info("Feature Label-Encoding abgeschlossen")

vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(X_text)
logging.info("TF-IDF-Vektorisierung abgeschlossen.")

X_combined = np.hstack([X_tfidf.toarray(), X_layer.values.reshape(-1, 1)])

X_train, X_test, y_train_feature, y_test_feature, X_layer_train, X_layer_test = train_test_split(
    X_tfidf, y_feature_encoded, X_layer, test_size=0.2,
    random_state=42, stratify=y_feature_encoded
)
logging.info(f"Percentage Split abgeschlossen. Trainingsgröße: {len(y_train_feature)}, Testgröße: {len(y_test_feature)}")

output_dir = "resampled_datasets_feature_prediction"
os.makedirs(output_dir, exist_ok=True)
logging.info(f"Hauptverzeichnis '{output_dir}' für Feature Resampling-Daten erstellt")

SAMPLERS = {
    "Adasyn": ADASYN(random_state=42),
    "SMOTE": SMOTE(random_state=42),
    "SMOTEENN": SMOTEENN(random_state=42),
    "Oversampling": RandomOverSampler(random_state=42),
    "Undersampling": RandomUnderSampler(random_state=42),
}

resampled_distribution = {}

for strategy, sampler in SAMPLERS.items():
    logging.info(f"Beginne Resampling mit Methode: {strategy}")

    X_train_resampled, y_train_feature_resampled = sampler.fit_resample(
        X_train, y_train_feature
    )

    logging.info(f"Resampling mit {strategy} abgeschlossen. Neue Größe: {len(y_train_feature_resampled)}")

    X_train_resampled_text = vectorizer.inverse_transform(X_train_resampled)

    resampled_df = pd.DataFrame({
        "Text": [" ".join(words) for words in X_train_resampled_text],
        "Predicted_Layer": np.tile(X_layer_train.values, len(y_train_feature_resampled) // len(y_train_feature) + 1)[:len(y_train_feature_resampled)],
        "Feature": feature_encoder.inverse_transform(y_train_feature_resampled)
    })

    strategy_folder = os.path.join(output_dir, strategy)
    os.makedirs(strategy_folder, exist_ok=True)
    file_path = os.path.join(strategy_folder, f"{strategy}_feature_resampled.csv")
    resampled_df.to_csv(file_path, index=False)
    logging.info(f"Resampled-Dataset für Feature gespeichert: {file_path}")

    resampled_distribution[strategy] = np.bincount(y_train_feature_resampled)

resampled_distribution_df = pd.DataFrame(resampled_distribution)
logging.info("DataFrame aus den gesammelten Resampling-Informationen wurde erstellt")

resampled_distribution_df.index = feature_encoder.classes_
resampled_distribution_path = os.path.join(output_dir, "resampled_class_distributions_feature.csv")
resampled_distribution_df.to_csv(resampled_distribution_path)
logging.info("Speicherung des DataFrames als .csv-Datei")

train_counts = np.bincount(y_train_feature)
test_counts = np.bincount(y_test_feature)

original_distribution_df = pd.DataFrame(
    {"Train": train_counts, "Test": test_counts},
    index=feature_encoder.classes_
)
original_distribution_path = os.path.join(output_dir, "original_train_test_distribution_feature.csv")
original_distribution_df.to_csv(original_distribution_path)
logging.info(f"Original Train-Test Verteilung gespeichert: {original_distribution_path}")
```

```python
import logging
import numpy as np
import pandas as pd
import torch
from pathlib import Path
from transformers import AutoModelForSequenceClassification, AutoTokenizer, Trainer, TrainingArguments
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import datasets
import os
from imblearn.over_sampling import ADASYN, SMOTE, RandomOverSampler
from imblearn.combine import SMOTEENN
from imblearn.under_sampling import RandomUnderSampler

os.environ["TOKENIZERS_PARALLELISM"] = "false"
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"

logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
logging.info(f"Using device: {device}")

LEARNING_RATES = [2e-5, 5e-5]
BATCH_SIZES = [8]
MODEL_NAME = "ClimateBERT/distilroberta-base-climate-f"
N_OUTER_SPLITS = 2
N_INNER_SPLITS = 2
N_EPOCHS = 2

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy_score(labels, predictions),
        "f1": f1_score(labels, predictions, average="macro"),
        "precision": precision_score(labels, predictions, average="macro"),
        "recall": recall_score(labels, predictions, average="macro"),
    }

from sklearn.preprocessing import LabelEncoder

feature_label_encoder = LabelEncoder()

def train_and_evaluate(train_texts, train_layers, train_labels, val_texts, val_layers, val_labels, model_name, learning_rate, batch_size, n_epochs):
    logging.info(f"Training with LR: {learning_rate}, Batch Size: {batch_size}")

    tokenizer = AutoTokenizer.from_pretrained(model_name)

    train_encodings = tokenizer(list(train_texts), truncation=True, padding=True, max_length=512)
    val_encodings = tokenizer(list(val_texts), truncation=True, padding=True, max_length=512)

    train_labels = feature_label_encoder.fit_transform(train_labels)
    val_labels = feature_label_encoder.transform(val_labels)

    train_labels = torch.tensor(train_labels, dtype=torch.long)
    val_labels = torch.tensor(val_labels, dtype=torch.long)

    train_dataset = datasets.Dataset.from_dict({
        "input_ids": train_encodings["input_ids"],
        "attention_mask": train_encodings["attention_mask"],
        "labels": train_labels.tolist()
    })

    val_dataset = datasets.Dataset.from_dict({
        "input_ids": val_encodings["input_ids"],
        "attention_mask": val_encodings["attention_mask"],
        "labels": val_labels.tolist()
    })

    num_labels = len(feature_label_encoder.classes_)
    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels)

    training_args = TrainingArguments(
        output_dir="./results",
        evaluation_strategy="epoch",
        per_device_train_batch_size=batch_size,
        per_device_eval_batch_size=batch_size,
        learning_rate=learning_rate,
        num_train_epochs=n_epochs,
        logging_dir="./logs",
        logging_steps=10,
        save_strategy="epoch",
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        compute_metrics=compute_metrics,
    )

    trainer.train()
    eval_results = trainer.evaluate()

    return eval_results

def process_strategy(strategy, output_folder, model_name):
    logging.info(f"Processing strategy: {strategy}")
    data_file = Path(output_folder) / strategy / f"{strategy}_feature_resampled.csv"

    if not data_file.exists():
        logging.warning(f"Data file not found for strategy {strategy}")
        return None

    data = pd.read_csv(data_file)
    texts, layers, labels = data['Text'], data['Predicted_Layer'], data['Feature']

    outer_cv = StratifiedKFold(n_splits=N_OUTER_SPLITS, shuffle=True, random_state=42)

    strategy_results = {
        "best_params": None,
        "best_results": None,
        "tested_combinations": []
    }

    best_f1 = 0

    for train_idx, test_idx in outer_cv.split(texts, labels):
        train_texts, test_texts = texts.iloc[train_idx], texts.iloc[test_idx]
        train_layers, test_layers = layers.iloc[train_idx], layers.iloc[test_idx]
        train_labels, test_labels = labels.iloc[train_idx], labels.iloc[test_idx]

        param_grid = {
            "learning_rate": LEARNING_RATES,
            "batch_size": BATCH_SIZES
        }

        for lr in LEARNING_RATES:
            for bs in BATCH_SIZES:
                inner_scores = []
                metrics_list = []

                inner_cv = StratifiedKFold(n_splits=N_INNER_SPLITS, shuffle=True, random_state=42)

                for inner_train_idx, val_idx in inner_cv.split(train_texts, train_labels):
                    inner_train_texts, val_texts = train_texts.iloc[inner_train_idx], train_texts.iloc[val_idx]
                    inner_train_layers, val_layers = train_layers.iloc[inner_train_idx], train_layers.iloc[val_idx]
                    inner_train_labels, val_labels = train_labels.iloc[inner_train_idx], train_labels.iloc[val_idx]

                    eval_results = train_and_evaluate(inner_train_texts, inner_train_layers, inner_train_labels,
                                                      val_texts, val_layers, val_labels,
                                                      model_name, lr, bs, N_EPOCHS)
                    inner_scores.append(eval_results.get("eval_f1", 0))
                    metrics_list.append(eval_results)

                if not metrics_list:
                    logging.warning("metrics_list is empty, skipping parameter evaluation.")
                    continue

                avg_metrics = {metric: np.mean([m[metric] for m in metrics_list]) for metric in metrics_list[0].keys()}

                strategy_results["tested_combinations"].append({
                    "learning_rate": lr,
                    "batch_size": bs,
                    "metrics": avg_metrics
                })

                if avg_metrics.get("eval_f1", 0) > best_f1:
                    best_f1 = avg_metrics["eval_f1"]
                    strategy_results["best_params"] = {"learning_rate": lr, "batch_size": bs}
                    strategy_results["best_results"] = avg_metrics

    logging.info(f"Best parameters for {strategy}: {strategy_results['best_params']}")

    results_path = Path("results_feature_prediction") / f"{strategy}_results.json"
    results_path.parent.mkdir(parents=True, exist_ok=True)

    pd.DataFrame(strategy_results["tested_combinations"]).to_json(results_path, orient="records", indent=4)

    return strategy_results

output_folder = Path("resampled_datasets_feature_prediction")

strategy = "Adasyn"
result_adasyn = process_strategy(strategy, output_folder, MODEL_NAME)


strategy = "SMOTE"
result_smote = process_strategy(strategy, output_folder, MODEL_NAME)

strategy = "SMOTEENN"
result_smoteenn = process_strategy(strategy, output_folder, MODEL_NAME)

strategy = "Oversampling"
result_oversampling = process_strategy(strategy, output_folder, MODEL_NAME)

strategy = "Undersampling"
result_undersampling = process_strategy(strategy, output_folder, MODEL_NAME)
```
