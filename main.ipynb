Der Code extrahiert Informationen mithilfe von regulären Ausdrücken (re) und zählt die Häufigkeiten mit Counter.

Da jede Zeile der Spalte Curation eine Liste enthält, ist die Aufgabe der Schleife, diese Listen zu durchlaufen und ihre Inhalte zu verarbeiten. Zunächst wird jede Liste in einen einzigen String umgewandelt, da reguläre Ausdrücke (re.findall) nur mit Strings arbeiten können. Dabei wird jedes Element der Liste mit der Funktion str() in einen String konvertiert, und anschließend werden die Elemente mithilfe von " ".join() zu einem String kombiniert.

Am Ende wird die Häufigkeit der gefundenen Layer, Features und Tags mit dem Counter-Modul analysiert, um ein besseres Verständnis der annotierten Daten zu erhalten.

```python
import re
from collections import Counter

layer_pattern = r"layer:([A-Za-z0-9_]+)"
feature_pattern = r"feature:([A-Za-z0-9_]+)"
tag_pattern = r"tag:([A-Za-z0-9_]+)"

sum_layers, sum_features, sum_tags = [], [], []

for curation_list in df['Curation']:
    curation_str = " ".join(map(str, curation_list))

    sum_layers.extend(re.findall(layer_pattern, curation_str))
    sum_features.extend(re.findall(feature_pattern, curation_str))
    sum_tags.extend(re.findall(tag_pattern, curation_str))

layer_distribution = Counter(sum_layers)
feature_distribution = Counter(sum_features)
tag_distribution = Counter(sum_tags)

print("Layer Distribution:", dict(layer_distribution))
print("Feature Distribution:", dict(feature_distribution))
print("Tag Distribution:", dict(tag_distribution))
```

Ungleichgewicht der Klassen (Imbalanced Data): Der Datensatz enthält Tags und Features, deren Häufigkeit stark variiert. Häufige Tags wie "Default addressee" dominieren, während seltene Tags wie "Quantitative target not mitigation" nur sporadisch auftreten.

Bei einem normalen K-Fold könnten einige Folds kaum Beispiele für die seltenen Klassen enthalten, was zu unzuverlässigen Testergebnissen führt.
Stratified K-Fold garantiert, dass die Verteilung der Klassen in jedem Fold proportional zur ursprünglichen Verteilung bleibt.

Granularität und Annotationen: Der Datensatz arbeitet auf Span-Ebene. Einzelne Annotationen könnten sich innerhalb eines Textes gegenseitig beeinflussen. Ein typischer Train/Test-Split oder normaler K-Fold könnte diese Abhängigkeiten ignorieren und zu einem Informationsleck führen.

Probleme bei K-Fold Cross-Validation: Keine Klassenbalance: Bei K-Fold wird der Datensatz ohne Rücksicht auf die Verteilung der Klassen aufgeteilt. Schwankende Ergebnisse: Die Modellleistung variiert stark zwischen den Folds, da die Klassenverteilung nicht repräsentativ ist. Dies führt zu einem ungenauen Bild der tatsächlichen Generalisierungsfähigkeit.

Probleme bei der typischen Datenaufteilung: Obwohl feste Splits effizient sind, können sie die Evaluierung verzerren, wenn der Test- oder Validierungssatz nicht repräsentativ für die Klassenverteilung ist.
Bei unausgeglichenen Datensätzen könnten seltene Klassen im Testset unterrepräsentiert sein, was zu optimistischeren Ergebnissen führt. Mit festen Splits werden Teile der Daten nur für das Testen reserviert und nicht für das Training genutzt. Das ist bei begrenzten Datenmengen nicht optimal.

Vorteile von Stratified K-Fold:
Erhält Klassenverteilungen: Jede Klasse ist in jedem Fold proportional zur Originalverteilung vertreten, wodurch die Ergebnisse fairer und repräsentativer werden.
Bessere Nutzung der Daten: Alle Datenpunkte werden in jedem Fold sowohl für das Training als auch für das Testen verwendet. Dies ist besonders vorteilhaft bei kleinen Datensätzen.
Robustere Ergebnisse: Durch die konsistente Verteilung der Klassen in jedem Fold sind die Evaluierungsergebnisse stabiler und weniger verzerrt.
Vermeidung von Überanpassung (Overfitting): Die Verteilung der Klassen über mehrere Folds minimiert das Risiko, dass das Modell eine bestimmte Klasse "überlernt", die nur in einem bestimmten Fold präsent ist.

Wir wählen Stratified K-Fold, weil es:
Die Herausforderung des Klassenungleichgewichts löst.
- Eine repräsentative Verteilung der Klassen in jedem Fold sicherstellt.
- Robustere und realistischere Ergebnisse liefert als K-Fold oder feste Splits.
- In deinem Datensatz mit stark variierenden Tag-Häufigkeiten ist Stratified K-Fold die ideale Wahl, um das Modell fair und effizient zu bewerten.

# ClimateBERT: Motivation und Methodik

## Motivation für die Entwicklung von ClimateBERT

1. **Herausforderungen in der Verarbeitung klimabezogener Fachsprache**:
   - Klimabezogene Texte zeichnen sich durch domänenspezifische Terminologie und komplexe Konzepte aus, die in allgemein vortrainierten Sprachmodellen (z. B. BERT) nicht hinreichend abgebildet werden können.
   - Konventionelle Ansätze wie Wörterbuchmethoden oder Bag-of-Words-Modelle sind unzureichend, da sie semantische und kontextuelle Nuancen nicht erfassen.

2. **Limitationen existierender Sprachmodelle**:
   - Modelle wie BERT oder ROBERTA basieren auf Textkorpora aus allgemeinen Domänen und zeigen suboptimale Leistungen bei der Verarbeitung von Nischensprachen wie klimabezogenen Texten.
   - Ohne domänenspezifisches Pretraining bleiben die Ergebnisse bei NLP-Aufgaben in der Klimaforschung eingeschränkt.

3. **Dringlichkeit in der Forschung und Praxis**:
   - Die wachsende Bedeutung von Klimaforschung erfordert robuste NLP-Modelle zur Extraktion und Analyse relevanter Informationen, etwa zur Klassifikation klimarelevanter Inhalte, zur Bewertung von Risiken und Chancen sowie zur Überprüfung von Fakten.
   - Das Fehlen eines spezialisierten Sprachmodells hat bisher die Effizienz und Präzision solcher Analysen begrenzt.

## Methodik hinter der Entwicklung von ClimateBERT

1. **Korpuserstellung**:
   - Eine umfassende Sammlung von über 2 Millionen Absätzen aus klimarelevanten Quellen, darunter Nachrichten, Forschungsabstracts und Unternehmensberichte.
   - Zielgerichtete Auswahl und Aufbereitung der Daten, um domänenspezifische Sprachmuster und Vokabulare abzubilden.

2. **Dreistufiger Trainingsansatz**:
   - **Generelles Pretraining**: Aufbauend auf DistilROBERTA, einem kompakten und effizienten Sprachmodell, das auf allgemeinen Textkorpora trainiert wurde.
   - **Domänenadaptives Pretraining**: Weitertraining auf einem klimarelevanten Korpus, um die Repräsentation von Fachterminologie und Sprachmustern zu verbessern.
   - **Aufgabenspezifisches Fine-Tuning**: Anpassung des Modells an konkrete klimabezogene NLP-Aufgaben wie Textklassifikation, Sentiment-Analyse und Faktenprüfung.

3. **Leistungssteigerungen durch ClimateBERT**:
   - Reduktion der Fehlerraten um bis zu 35,71 % und signifikante Verbesserung der F1-Scores bei verschiedenen klimabezogenen Aufgaben.
   - Überlegenheit gegenüber allgemeinen Modellen wie DistilROBERTA in allen getesteten Szenarien.

4. **Berücksichtigung von Nachhaltigkeitsaspekten**:
   - Verwendung energieeffizienter Modellarchitekturen und Minimierung des Ressourcenverbrauchs durch gezielte Datenverarbeitung und begrenztes Hyperparameter-Tuning.
   - Kompensation des CO₂-Fußabdrucks durch Spenden an anerkannte Klimaschutzorganisationen.

# Masked Language Model (MLM)

## Definition
Das Masked Language Model (MLM) ist eine Trainingsmethode, bei der Sprachmodelle durch das Vorhersagen maskierter Wörter im Text lernen, den Kontext zu verstehen. Es wird vor allem in Modellen wie **BERT** und seinen Varianten eingesetzt.

---

## Funktionsweise
1. **Maskierung**:
   - Zufällig ausgewählte Tokens (Wörter) im Eingabetext werden durch ein spezielles `[MASK]`-Token ersetzt.
   - Beispiel:
     - Originaltext: „Das Wetter ist heute sonnig.“
     - Maskierter Text: „Das Wetter ist [MASK] sonnig.“

2. **Vorhersage**:
   - Das Modell versucht, die maskierten Tokens anhand des Kontexts zu rekonstruieren.
   - Beispiel: Das Modell sollte „heute“ als Vorhersage für `[MASK]` liefern.

3. **Loss-Berechnung**:
   - Die Differenz zwischen den tatsächlichen Wörtern (Labels) und den vom Modell vorhergesagten Wörtern wird berechnet.
   - Häufig wird **Cross-Entropy Loss** verwendet, um die Wahrscheinlichkeit der korrekten Vorhersagen zu maximieren.

---

## Vorteile
- **Kontextsensitives Lernen**: Das Modell berücksichtigt den gesamten Kontext (links und rechts der Maske), um tiefere Sprachmuster zu lernen.
- **Flexibilität**: Ermöglicht die Entwicklung allgemeiner Sprachrepräsentationen, die in verschiedenen NLP-Aufgaben (z. B. Textklassifikation, Sentiment-Analyse) verwendet werden können.

---

## Beispiel
### Eingabe:
„Die Auswirkungen des [MASK] Wandels sind gravierend.“

### Ziel:
Das Modell soll `[MASK]` korrekt mit „Klimawandels“ ersetzen.

## 1. Textklassifikation

  - ClimateBERT reduziert den **Cross-Entropy-Loss** um **32,64 %** im Vergleich zu DistilROBERTA.
    - **Cross-Entropy-Loss**: Eine Maßzahl, die angibt, wie gut die vom Modell vorhergesagten Wahrscheinlichkeiten mit den tatsächlichen Zielwerten übereinstimmen. Niedrigere Werte bedeuten bessere Vorhersagen.
  - Die Fehlerquote beim **F1-Score** wird um **35,71 %** gesenkt.

## 2. Sentiment-Analyse

  - **Neutral**: Keine Wertung oder Bewertung.
- **Daten**:

## Änderungen am Code

### Berücksichtigung von Klassengewichten

#### Ursprünglich
Der ursprüngliche Code berücksichtigt keine **Class Weights**, was bei unausgeglichenen Datensätzen zu einer schlechteren Modellleistung für unterrepräsentierte Klassen führt.

#### Neu
**Class Weights** werden mit `compute_class_weight` berechnet und in die Verlustfunktion integriert. Dadurch werden unterrepräsentierte Klassen stärker gewichtet.

### Benutzerdefinierte Modellklasse (WeightedModel)

#### Ursprünglich
- Verwendet direkt das Modell `AutoModelForSequenceClassification` ohne Änderungen.
- Die Verlustfunktion ist standardmäßig ohne Anpassung an **Klassengewichte**.

#### Neu
- Ein benutzerdefiniertes Modell (`WeightedModel`) umschließt das Basismodell.
- **Class Weights** werden direkt in die Verlustfunktion integriert.

### Gerät und Datentyp-Kompatibilität

#### Ursprünglich
- Es fehlt eine explizite Überprüfung, ob **Labels** und **Logits** auf demselben Gerät (z. B. GPU) sind und kompatible Datentypen verwenden.

#### Neu
- **Labels** und **Logits** werden in der `forward`-Methode des benutzerdefinierten Modells explizit auf `dtype=torch.long` und das gleiche Gerät wie die **Logits** gebracht.

### Verbesserte Behandlung von Warnungen in `compute_metrics`

#### Ursprünglich
- Es gab keine Absicherung gegen Division durch Null in `precision_recall_fscore_support`.

#### Neu
- Das Argument `zero_division=0` wurde hinzugefügt, um Berechnungsprobleme zu vermeiden.

- Der `Trainer` wurde direkt mit dem Standardmodell und der Standard-Verlustfunktion verwendet.

#### Neu
- Der `Trainer` verwendet das benutzerdefinierte Modell mit **Klassengewichten**, um den unausgeglichenen Datensatz auszugleichen.

- Die Standardverlustfunktion wurde verwendet, ohne Anpassung an **Klassengewichte**.

### Warum wurde der Datensatz geschaffen?

Allerdings gibt es beträchtliche Unsicherheiten darüber, wie die Wirksamkeit dieser Maßnahmen in prominenten Analysen gemessen werden kann. Beispielsweise basieren einige Studien ihre Analysen auf der „Dichte“ von Politiken (d. h. der Anzahl von Maßnahmen), während andere sich auf die von diesen Maßnahmen gesetzten Ziele oder Vorgaben konzentrieren.

„Policy-Instrumententypen“, die eine übergeordnete Interventionslogik implizieren (z. B. „Regulierung“, um bestimmte Handlungen zu erlauben oder zu verbieten).

„Allgemeine Gestaltungseigenschaften“, die grundlegende Designentscheidungen jeder Politik umfassen 

„Spezifische Gestaltungseigenschaften“, die Designentscheidungen betreffen, die stärker auf ein bestimmtes Politikfeld zugeschnitten sind.

Die Wahl des Instrumententyps ist wichtig, jedoch besteht ein wachsender Konsens, dass kein Instrumententyp bei der Minderung des Klimawandels allen anderen überlegen ist.

Jede Politik umfasst eine Reihe allgemeiner Gestaltungselemente, unabhängig von den gewählten Instrumententypen. So konzeptualisieren beispielsweise Cashore und Howlett Ziele und Mittel auf unterschiedlichen Abstraktionsebenen, während Schaffrin et al. sechs allgemeine Gestaltungselemente vorschlagen: Ziele, Umfang, Integration, Budget, Implementierung und Überwachung. Diese allgemeinen Gestaltungseigenschaften sind für die Wirksamkeit von Politiken mindestens ebenso wichtig wie die Wahl des Instrumententyps.

Während bereits Anstrengungen unternommen wurden, klimabezogene Politikdaten zu sammeln, konzentrieren sich bestehende Datenbanken wie LSEs „Climate Change Laws of the World“ und IEAs „Policies and Measures“ auf allgemeine Informationen (z. B. Umfang, Themen, abgedeckte Sektoren, Instrumententypen) und nicht auf spezifischere Bewertungen der Gestaltungselemente von Politiken.

Da das Hochfahren der Ambitionen in der Klimapolitik auf den bestmöglichen verfügbaren Daten basieren sollte, schränkt der Mangel an Daten zu den meisten Gestaltungselementen von Politiken die systematische Bewertung der Ambition und Wirksamkeit von Klimapolitiken ein.

Da das Hochfahren der Ambitionen in der Klimapolitik auf den bestmöglichen verfügbaren Daten basieren sollte, schränkt der Mangel an Daten zu den meisten Gestaltungselementen von Politiken die systematische Bewertung der Ambition und Wirksamkeit von Klimapolitiken ein. Daher sind Alternativen zur manuellen Kodierung von Politiken dringend erforderlich.

Beispielsweise haben jüngste Studien Ansätze, die Texte als Daten verwenden, eingesetzt, um einzelne Aspekte von Politikdesigns zu identifizieren, wie deren Thema oder Umfang (z. B. Anpassung an den Klimawandel). In verwandten Bereichen wie der Forst- und Umweltpolitik haben Forscher solche Ansätze genutzt, um die verwendeten Instrumente (d. h. Instrumententypen) und formulierten Ziele zu identifizieren. Weitere Ansätze zur Bewertung allgemeiner und spezifischer Gestaltungselemente von Politiken im großen Maßstab sind jedoch bisher nicht entstanden.

In Ermangelung solcher Ansätze basieren Analysen der Wirksamkeit von Politiken weiterhin auf manuell kodierten Daten. Hier liefern wir einen ersten Schritt in Richtung der Erstellung groß angelegter Klimapolitik-Datensätze mithilfe von textbasierten Ansätzen.

Um das langfristige Ziel zu erreichen, die Bewertung von Gestaltungselementen der Klimapolitik (teilweise) zu automatisieren und damit die ressourcenintensive manuelle Kodierung zu unterstützen oder zu ersetzen, stellt dieses Paper einen Trainingsdatensatz vor, der zur Entwicklung neuer Analysewerkzeuge dienen kann. Der Datensatz erweitert den Anwendungsbereich textbasierter Ansätze in der Politikanalyse erheblich, da er eine umfassende Reihe von Gestaltungselementen auf verschiedenen Abstraktionsebenen umfasst.

Der POLIANNA-Datensatz soll die Arbeit von Politikforschern bei der Analyse der Ambitionen und Wirksamkeit von Klimapolitik auf verschiedenen Ebenen unterstützen. Die auf diesem Datensatz basierenden zukünftigen Identifikationswerkzeuge könnten dabei helfen, viele der zentralen Fragen im Zusammenhang mit dem dringend benötigten Hochfahren der Klimapolitikambitionen zu bearbeiten, wie beispielsweise die Wirksamkeit von Politiken, das potenzielle Zusammenspiel zwischen verschiedenen Politiken und Wege zur Verbesserung des Politikdesigns im Laufe der Zeit. Dies ist besonders relevant für die Entwicklung evidenzbasierter Politikempfehlungen.

Da empirische Large-N-Studien zur Analyse der Wirksamkeit von Politiken idealerweise Design-Daten von mehreren hundert Politiken erfordern würden, sind die Kosten für eine vollständige manuelle Annotation prohibitiv hoch. Die hohe Zeit- und Ressourcenbelastung zeigt den dringenden Bedarf an:

Automatisierung der Annotation:
Entwicklung von (teil-)automatisierten Werkzeugen, die den Annotatoren helfen oder den Prozess teilweise ersetzen können.
Skalierbarkeit:
Nur durch skalierbare Ansätze wie maschinelles Lernen oder NLP-Methoden kann eine systematische Analyse großer Politikdatensätze realistisch umgesetzt werden.

### Wie wurde er erstellt?

Die Datensammlung basiert auf manueller Kodierung, die von geschultem Personal durchgeführt wird, oft in Zusammenarbeit mit Länderspezialisten oder Regierungsbeamten. Dieser arbeitsintensive Ansatz ist nur schwer zu skalieren, nicht zuletzt, weil die Finanzierung der sozialwissenschaftlichen Forschung zu Klimawandelthemen begrenzt ist.

Zu diesem Zweck haben wir ein Kodierungsschema entwickelt, das diese Gestaltungselemente widerspiegelt und maschinell lernbare Aufgaben definiert. Um den Datensatz der Politikanmerkungen (POLIANNA) zu erstellen, wählten wir einen Korpus relevanter EU-Klimapolitiken (d. h. Richtlinien und Verordnungen, die sich auf die Minderung des Klimawandels und erneuerbare Energien beziehen) aus. Wir entwickelten ein Schema zur Annotation von Instrumententypen, Gestaltungseigenschaften von Politiken sowie der Technologie- und Anwendungsspezifität (im Folgenden der Kürze halber nur als Technologiespezifität bezeichnet) auf der Ebene von Textabschnitten. Wir entschieden uns, auf einer textuell granularen Ebene zu annotieren, da dies der Art und Weise, wie menschliche Annotatoren eine Politik analysieren, näher kommt und gleichzeitig mehr Flexibilität für die Entwicklung unterschiedlicher Analysewerkzeuge bietet. Anschließend kodierten wir mit einem Team von geschulten Annotatoren manuell einen umfangreichen Textkorpus.

In diesem komplexen Kontext analysierten wir die Übereinstimmung zwischen den Annotatoren ausführlich, passten einen chancekorrigierten Score an und entwickelten weitere Übereinstimmungsmaße basierend auf dem F1-Score

Da einige Informationen zum Politikdesign kontext- und domänenspezifisch sind, wird es voraussichtlich weiterhin schwierig sein, diese vollständig zu automatisieren. Aus diesem Grund haben wir die Texte auf der granularsten Ebene annotiert, um die größtmögliche Flexibilität bei der Entwicklung solcher Werkzeuge zu ermöglichen.

Zusammenstellung eines Korpus von EU-Politiken, d. h. Richtlinien und Verordnungen, die sich auf den Klimawandel und erneuerbare Energien beziehen.
Vorverarbeitung der Texte, um sie für die Annotation vorzubereiten.
Umfassendes Testen und Überarbeiten des Kodierungsschemas, gefolgt von der Annotation eines großen Textkorpus durch ein Team geschulter Annotatoren (oder „Kodierer“) und der abschließenden Kuratierung der Annotationen für den finalen Datensatz. 

Quellen der Politiktexte
Die Politiktexte wurden von EUR-Lex, der offiziellen Website für den Zugang zu Rechtsdokumenten der Europäischen Union, bezogen. Diese Plattform bietet Open-Source-Zugang zu allen EU-Rechtsakten, einschließlich Verordnungen, Richtlinien, Entscheidungen, Empfehlungen und Stellungnahmen, die in unserer Studie als EU-„Politiken“ betrachtet werden. EUR-Lex stellt eine „Experten-Suchfunktion“ zur Verfügung, die es ermöglicht, gezielte Suchanfragen für die Identifizierung relevanter Dokumente durchzuführen.

In einem iterativen Prozess entwickelten wir Suchanfragen, die bewusst breit angelegt waren, um sowohl Politiken zur allgemeinen Minderung des Klimawandels als auch Maßnahmen zu spezifischen Minderungsaktionen (z. B. Technologien zur Dekarbonisierung relevanter Sektoren) zu erfassen.Für die Suche nutzten wir den NEAR10-Operator, um eine breite Palette von Politiken einzufangen, darunter Kombinationen von Begriffen wie „erneuerbare Energien“ und/oder „Klimawandel“. Diese breite Erfassung wurde später manuell eingegrenzt.

Aus dem breiten Katalog der durch die Suchanfragen gefundenen Politiken wählten wir eine kleinere Anzahl von Texten für die Annotation aus. Zunächst sortierten wir die Texte nach dem Prozentsatz der Artikel, die Schlüsselwörter zu Energie und Klimaschutz enthielten. Diese Schlüsselwörter wurden aus den ursprünglichen Suchanfragen abgeleitet, verfeinert und mithilfe regulärer Ausdrücke identifiziert. Anschließend wählten wir die zehn wichtigsten Richtlinien und Verordnungen aus, schlossen jedoch Entscheidungen aufgrund ihres begrenzten Anwendungsbereichs aus. Verordnungen müssen in ihrer Gesamtheit in der gesamten EU angewendet werden, Richtlinien setzen Ziele, die die Mitgliedstaaten auf eigene Weise umsetzen müssen, während Entscheidungen spezifische Fälle betreffen und nur für die Adressaten bindend sind. Die Metadaten enthalten eindeutige Kennungen, die angeben, ob ein Text zu den Kategorien „Verordnung“ oder „Richtlinie“ gehört.Zusätzlich wählten wir acht weitere Politiken basierend auf ihrem Titel und ihren Schlüsselwörtern aus, um zusätzliche Anforderungen an den Korpus zu erfüllen. Ziel war es, eine Vielzahl von Minderungsmaßnahmen und wirtschaftlichen Sektoren abzudecken. Beispielsweise wurden Politiken einbezogen, die sich spezifisch mit Technologien zur Energiespeicherung (z. B. Batteriespeicher) oder dem Verkehrssektor (z. B. Dekarbonisierung des Transports durch den Einsatz von Elektrofahrzeugen) befassen.

Da einige der Politiktexte sehr lang sind, haben wir die Dokumente in kürzere Textentitäten unterteilt, nämlich die Artikel, wie sie in den EU-Rechtsdokumenten spezifiziert sind. Diese Aufteilung brachte mehrere Vorteile:

Vergleichbarkeit: Die Textentitäten waren in ihrer Länge besser vergleichbar.
Effizienz: Die Zuweisung der Texte an Annotatoren wurde erleichtert.
Wir annotierten jeden Artikel jeder Politik mit Ausnahme der Präambel, die Zitate und Erwägungsgründe umfasst. Diese wurde aus dem annotierten Korpus ausgeschlossen, da die rechtliche Grundlage oder verfahrensbezogene Informationen nicht im Fokus unserer Analyse lagen.

Die Skripte zur Aufteilung von EU-Politiken in Artikel sind online verfügbar. Insgesamt teilten wir die 18 Politiken in 448 Artikel auf, reduzierten diese jedoch auf 412, indem wir Zitate und Erwägungsgründe ausschlossen.

Wir haben ein umfassendes Kodierungsschema entwickelt, um drei Gruppen von Gestaltungselementen politischer Maßnahmen zu erfassen:

Instrumententypen
Gestaltungseigenschaften der Politik
Technologiespezifität

Das Kodierungsschema ist so gestaltet, dass es auf der Ebene von Textspannen angewendet werden kann, die aus einem oder mehreren Wörtern (Tokens) bestehen.

Für die Annotationen verwendeten wir das Open-Source-Tool Inception der TU Darmstadt. Dieses Werkzeug erfüllte alle Anforderungen für die Annotationen, darunter:

Annotationen auf Span-Ebene,
überlappende Annotationen,
Unterstützung mehrerer Annotatoren,
eine hierarchische Labelstruktur.

Inception ermöglichte uns die Definition einer dreistufigen hierarchischen Struktur:

Oberste Ebene (Layer):
Drei „Layer“ entsprechen den oben eingeführten Gestaltungselementen:

Instrumententypen,
Gestaltungseigenschaften,
Technologiespezifität.
Mittlere Ebene (Features):
Innerhalb jedes Layers definierten wir mehrere Features, die im Text identifiziert werden sollten.

Unterste Ebene (Tags):
Für jedes Feature konnten Annotatoren aus einer Reihe von Tags wählen, um eine Textspanne zu annotieren.

Annotation auf der Spannen-Ebene vorzunehmen, um maximale Flexibilität für zukünftige Vorhersagetools zu gewährleisten, die auf diesen granularen Daten basieren. Diese Entscheidung bietet mehrere Vorteile:

Aggregierbarkeit: Labels auf der Spannen-Ebene können einfach auf Satz-, Absatz- oder Dokumentenebene aggregiert werden.
Semi-automatisierte Tools: Da die Verarbeitung und Klassifizierung bestimmter Gestaltungselemente ein hohes Maß an Textverständnis und Fachwissen erfordert, erwarten wir, dass viele Elemente nur mit halbautomatisierten Tools und menschlicher Unterstützung zufriedenstellend bearbeitet werden können.
Modelltraining: Annotationen auf Spannen-Ebene ermöglichen das Training von Modellen, die relevante Passagen im Text markieren und interpretierbare Informationen für Menschen bereitstellen können, die die Annotation überwachen.

Während die meisten Tags wechselseitig exklusiv sind, wurde das Akteur-Tag-Set bewusst anders gestaltet. Es enthält eine implizite vierte hierarchische Ebene, in der die Tags in zwei Hauptkategorien unterteilt sind:
Adressat (addressee): Akteure, die Ziel der politischen Maßnahme sind.
Verantwortliche Instanz (authority): Akteure, die für die Umsetzung verantwortlich sind.

Standardkategorie: Da diese Unterkategorien schwer zu kodieren sein können, haben wir eine Standardkategorie eingeführt, die angewendet werden kann, wenn spezifischere Kategorien nicht zugeordnet werden können.

Die Annotatoren mussten zwei Hauptaufgaben bewältigen:

Festlegung der Spannen-Grenzen (Unitising): Bestimmung der exakten Textpassage, die relevant ist.
Kategorisierung: Zuweisung der passenden Labels zu jeder Spanne.

Embedding (hierarchische Überlappung): Eine Spanne kann in einer anderen verschachtelt sein.
Freie Überlappung: Spannen können sich überlappen, ohne hierarchisch verschachtelt zu sein.
Sporadicity: Nicht zusammenhängende Spannen können als Teil derselben Annotation betrachtet werden.
Die Länge der Spannen wurde nicht im Voraus festgelegt. Die Annotatoren wurden jedoch angewiesen, die kürzest mögliche Spanne zu wählen, die noch aussagekräftig ist.

Die Inter-Annotator Agreement (IAA) misst die Übereinstimmung zwischen Annotatoren, ein zentraler Bewertungsmaßstab in der computerlinguistischen und textbasierten Datenanalyse. Diese Übereinstimmung wird oft als Inter-Coder Reliability, Inter-Coder Agreement, Inter-Rater Reliability oder unter ähnlichen Begriffen bezeichnet.

Bedeutung der IAA
Bewertung der Zuverlässigkeit:

Die IAA zeigt, wie stark die Annotatoren bei derselben Aufgabe übereinstimmen.
Eine niedrige Übereinstimmung kann darauf hinweisen, dass:
Die Annotatoren nicht ausreichend geschult sind, oder
Die Aufgabe und das Kodierungsschema nicht klar definiert sind.

### Wer hat ihn erstellt?

Die Daten wurden von vier Autoren annotiert, die über relevante Fachkenntnisse auf Master-Niveau in Politikwissenschaft sowie erneuerbarer Energietechnik und -politik verfügten. Der Trainingsprozess umfasste:

Intensive Schulung: Mehrere Dutzend Meetings zur Einführung in das Kodierungsschema und die Annotation.
Testläufe: Zahlreiche Probe-Annotationen, die zur Entwicklung des finalen Kodierungsschemas führten.
Diskussion von Grenzfällen: Alle während des Trainings aufgetretenen Sonderfälle wurden umfassend besprochen.
Da die Aufgabe ein hohes Maß an fachlicher Expertise erforderte, war der Schulungsprozess intensiv.

Doppelte Annotation: Der gesamte Korpus wurde unabhängig von zwei Annotatoren annotiert.
Kuratierung:
Eine der korrespondierenden Autoren oder, in wenigen Fällen, ein Annotator unter ihrer Aufsicht kuratierte die Annotationen.
Der Kurator aggregierte die Labels beider Annotatoren, löste Konflikte und nahm, falls nötig, Änderungen vor, wie z. B.:
Hinzufügen neuer Labels,
Entfernen fehlerhafter Labels.
Dieser iterative und qualitätsgesicherte Prozess stellte sicher, dass die Annotationen sowohl präzise als auch zuverlässig sind.

Unitisation:
Annotatoren müssen die Position und Länge der Spanne (ein oder mehrere Tokens) bestimmen.
Kategorisierung:
Annotatoren müssen die korrekten Tags auswählen.
Überlappende Annotationen:
Wir erlauben Annotationen, die sich überlappen können, was die Komplexität erhöht.

### Wie arbeitet man mit dem Datensatz?

Potenzielle NLP-Aufgaben
Named Entity Recognition (NER): Identifikation bestimmter Entitäten in Textspannen.
Satzklassifikation: Wenn die Annotation auf Satzebene aggregiert wird.
Dokumentklassifikation: Wenn die Annotation auf Absatz- oder Dokumentebene aggregiert wird.

Für Analysen, bei denen die Unterkategorien nicht relevant sind, empfehlen wir eine Aggregation auf die Hauptkategorien „Adressat“ und „Verantwortliche Instanz“, um die Komplexität zu reduzieren und die Vergleichbarkeit zu erleichtern.

Implikationen für maschinelles Lernen
Die Verteilung der Spannenlängen bietet wichtige Erkenntnisse für die Entwicklung von ML-Modellen:

Named Entity Recognition (NER):
Aufgrund der kurzen Länge vieler Spannen ist NER besonders geeignet, um die meisten Tags vorherzusagen.
Satzklassifikation:
Für Tags, die längere Annotationen erfordern, wie etwa "objectives", könnte die Satzklassifikation hilfreich sein.
Effizienz:
Die kurze Länge der meisten Spannen reduziert die Komplexität der Annotation, was die Modellentwicklung vereinfacht.
Die Analyse der Spannenlängen zeigt, dass ML-Ansätze, die auf spezifische NLP-Aufgaben wie NER oder Satzklassifikation ausgerichtet sind, ein hohes Potenzial haben, um die Annotationen effektiv zu unterstützen.

Benchmark für ML-Modelle:

Die IAA dient als Vergleichsmaßstab für die Leistung automatisierter Annotationen (z. B. durch ML-Modelle).
Da ML-Modelle typischerweise nicht mit zufallskorrigierten Scores bewertet werden, nutzen wir beide:
Zufallskorrigierte IAA-Scores, um Annotator-Leistungen zu messen.
Standard-ML-Metriken, wie den F1-Score, zur Bewertung von ML-Modellen.

F1-Score:
Grundlage:
Einer der Annotatoren wird als Ground Truth definiert.
Der andere Annotator wird als Vorhersage betrachtet.

Da der F1-Score nicht symmetrisch ist, wird die Berechnung zweimal durchgeführt:
Mit Annotator A als Ground Truth und Annotator B als Vorhersage.
Mit Annotator B als Ground Truth und Annotator A als Vorhersage.
Der endgültige F1-Score ist der Durchschnitt der beiden Berechnungen.

Kuratierung als Ground Truth:
Um die Übereinstimmung der Annotatoren mit der kuratierten Version zu bewerten, wird die Kuratierung als Ground Truth betrachtet.
Für jeden Annotator wird der F1-Score berechnet, und der Durchschnitt der beiden Werte wird gebildet.

Precision: Anteil der korrekt markierten Annotationen (True Positives) an allen markierten Annotationen.
Recall: Anteil der korrekt markierten Annotationen (True Positives) an allen erwarteten Annotationen.

Der F1 Exact-Score ist eine strikte Metrik, die Annotationen nur dann als übereinstimmend wertet, wenn sie in beiden Dimensionen exakt übereinstimmen:

Unitisation:
Die Spannen müssen exakt übereinstimmen (Start- und Endposition müssen identisch sein).

Kategorisierung:
Die annotierten Tags müssen ebenfalls übereinstimmen.

Der F1 Exact-Score ist eine sehr strikte Metrik, da sie keine Abweichungen erlaubt. Dies führt zu:

Strafen bei hierarchischen Überlappungen:

Beispiel: Ein Annotator markiert die Spanne „battery and storage technology“ sowie die Teilspannen „battery“ und „storage technology“ mit dem Tag Tech_LowCarbon.
Der zweite Annotator markiert nur die Spanne „battery and storage technology“.
Der F1 Exact-Score zählt die zusätzlichen Annotationen des ersten Annotators als False Positives, obwohl beide Ansätze akzeptabel sind.
Einschränkungen bei überlappenden Annotationen:

Unser Kodierungsschema erlaubt und fördert in bestimmten Fällen überlappende Annotationen.
Der F1 Exact-Score kann diese Besonderheit nicht abbilden und führt zu Strafen, selbst wenn beide Annotatoren inhaltlich korrekt gearbeitet haben.

Um die Einschränkungen des F1 Exact-Score zu adressieren, haben wir den F1 Heuristic-Score eingeführt. Der F1 Heuristic-Score ist eine weniger strenge Metrik zur Bewertung der Übereinstimmung zwischen Annotatoren oder zwischen Annotator und Kurator. Sie berücksichtigt überlappende Annotationen und erlaubt flexiblere Zuordnungen, wodurch sie besonders für komplexe Szenarien wie hierarchische oder granulare Annotationen geeignet ist.

Eine Annotation wird als Übereinstimmung gezählt, wenn:

Die Kategorisierung (Tag) übereinstimmt.
Es eine Überlappung zwischen den Spannen gibt, unabhängig davon, ob diese exakt übereinstimmen.

Für jede Kategorie (Tag) wird definiert:

True Positives (tp): Anzahl der Spannen in Annotator A, die eine Kategorisierungsübereinstimmung und eine Überlappung mit einer Spanne in der Ground Truth (GT) aufweisen. Mehrere Tags in A können derselben Tag in GT zugeordnet werden.
False Positives (fp): Anzahl der Spannen in Annotator A, die keine überlappende Spanne mit Kategorisierungsübereinstimmung in GT haben.
False Negatives (fn): Anzahl der Spannen in GT, die keine überlappende Spanne mit Kategorisierungsübereinstimmung in Annotator A haben.

Der F1 Token-wise-Score misst die Übereinstimmung zwischen Annotatoren oder zwischen Annotator und Kurator, indem er die Anzahl der übereinstimmenden Tokens berücksichtigt. Im Gegensatz zu anderen Metriken wird hier keine Unitisation (d. h. Start- und Endposition von Spannen) bewertet. Stattdessen liegt der Fokus darauf, wie viel Text in beiden Annotationen mit derselben Kategorie markiert wurde.

Keine Unitisation:

Es wird nicht geprüft, ob die Start- und Endpositionen der Spannen exakt übereinstimmen.
Nur die Übereinstimmung der Kategorien und Tokens zählt.
Berücksichtigung von Überlappungen:

Die Regeln für überlappende Annotationen sind dieselben wie beim F1 Heuristic-Score.
Wenn zwei Annotationen sich überschneiden und dieselbe Kategorie haben, werden die Tokens, die in beiden enthalten sind, gezählt.
Gewichtung längerer Spannen:

Längere Spannen erhalten automatisch ein höheres Gewicht, da mehr Tokens übereinstimmen können.
Dies macht die Metrik besonders nützlich, wenn längere Annotationen häufiger vorkommen oder bedeutsamer sind.

Vorteile des F1 Token-wise-Score
Textabdeckung:
Misst, wie viel des Textes korrekt kategorisiert wurde, unabhängig von der genauen Position der Spannen.
Gewichtung längerer Spannen:
Längere Spannen mit hoher Übereinstimmung beeinflussen den Score stärker, was sinnvoll ist, wenn längere Annotationen bedeutsamer sind.
Flexibilität bei Überlappungen:
Überlappende Spannen werden berücksichtigt, ohne sie strikt zu bestrafen.

Gamma (y) - Score:
Der γ-Score ist ein zufallskorrigiertes Maß zur Bewertung der Inter-Annotator Agreement (IAA), das speziell für komplexe Annotationen entwickelt wurde, die Unitisation (Bestimmung von Position und Länge), Kategorisierung (Zuweisung von Tags) und Überlappung (overlap) umfassen.
Einheitliches Maß:
Der γ-Score bewertet die Übereinstimmung zwischen Annotatoren und richtet gleichzeitig Annotationen zueinander aus.

Chance-Korrektur:
Der γ-Algorithmus berechnet einen Wert für alle lokalen Unterschiede zwischen Annotatoren (sogenannte „Disorder“).
Die minimale „Disorder“, also der bestmögliche Abgleich der Annotationen, wird korrigiert, indem die „erwartete Disorder“ aus zufällig neu gemischten Annotationen ermittelt wird.

Integration von Unitisation und Kategorisierung:
Der γ-Score berücksichtigt sowohl:

Positionierung: Wie gut die Annotatoren die Grenzen der Spannen übereinstimmend markieren.
Kategorisierung: Wie gut die Tags der Annotatoren übereinstimmen.

Weiche Penalties für Tag-Unterschiede:

Standardmäßig wird jeder Unterschied zwischen Tags mit einer Strafe von 1 belegt.

Wertebereich:

Der γ-Score liegt zwischen 1 (perfekte Übereinstimmung) und 0 (keine Übereinstimmung).
Durch die Möglichkeit, die Strafen für Tag-Unterschiede anzupassen, ist der γ-Score flexibler als starre Bewertungsmetriken.
